```python
import pandas as pd
import sqlite3

# Caminho do arquivo CSV
csv_file_path = '/content/drive/MyDrive/Codes/unimed/base_dados_teste_analista.csv'  # Substitua pelo caminho do seu arquivo CSV
db_file_path = '/content/drive/MyDrive/Codes/unimed/base_dados_teste_analista.db'      # Nome do banco de dados SQLite que será criado
table_name = 'vendas_2023'               # Nome da tabela no banco de dados

# Carregar o CSV em um DataFrame
df = pd.read_csv(csv_file_path)

# Conectar ao banco de dados SQLite (ou criar um novo)
conn = sqlite3.connect(db_file_path)

# Salvar o DataFrame no banco de dados
df.to_sql(table_name, conn, if_exists='replace', index=False)

# Fechar a conexão com o banco de dados
conn.close()

print(f"Arquivo CSV convertido para o banco de dados SQLite: {db_file_path}")
```
